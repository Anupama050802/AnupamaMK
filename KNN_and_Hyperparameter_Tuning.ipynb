# KNN(K Nearest Neighbours)

- Simple ML algorithm that predicts the label of a new sample based on the K closest neighbours. K any no. like n
- Lazy learner---> there is no model training, it stores just training data and tells based on that.
- Instance based learner---> it uses the actual stored data points to make predictions directly without undergoing any training it just learns and keep by seeing its neighbours as it is.
- Non parametric---> no assumptions about the shape of the data.(like bell shaped-min max etc no botheration)

## What type of ML algorithm it is?
- supervised
- can be used for both regression and classification
- because the output is decided based on the neighbours: if we check:
  - majority vote---> classification
  - average of the neighbours---> regression

## Where is KNN used?
- data is small or medium sized where storing the data is not expensive as it does not use model instead stores the data as it is. around 500 rows
- whenever decision boundary is irregular or non linear ( logistic regression fails here as logistic is smooth curve and KNN can adapt locally)

## How does KNN actually work?

for any new sample/test data:

Step 1: Calculate the Euclidean distance between the new sample point and ALL training samples.

Step 2: Arrange the distance and pick the K nearest neighbours to that sample point.K should be an odd value:3,5,7
Because if 4 and 2 says one and other 2 says something we cant come to conclusion

Step 3:
- Classification---> majority vote
- Regression---> average of neighbour values

Step 4:
- Assign the predicted output.
- No "learn" function.It waits untill prediction to take further action.





## Implementation

# Step 1: Problem statement

```python
#Predict whether the tumor is benign or malignant
```

# Step 2: Load Dataset

```python
# load the dataset from sklearn
# convert df
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data= load_breast_cancer(as_frame=True)
df=data.frame # to get even target
df.head()
```

# Step 3: Preprocessing

```python
# separate the features and the target
x= df.drop('target',axis=1) # all columns except the target
y= df['target']
```

```python
# scaling using standard scaler can be done
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x) # scaled version of features
```

# Step 4: Train-Test Split

```python
##                 x(features)                 y(output)
##
##Training(70%)-->   x_train                   y_train
##Testing(30%)-->    x_test                    y_test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42) # order of variables like this important
```

```python
# shape
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Step 5: Train the model

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=5) #k=3,5,7,so on 7 no need as might be widely spread and then wont be nearest

knn.fit(x_train,y_train) # no as such training it its just rough training calculating distance and fininding neighbours and testing for it
```

# Step 6: Evaluate the model

```python
# prediction # based on that we predict and see all evaluations how it has come based on that we can change
y_pred= knn.predict(x_test)
```

```python
# calculate accuracy,precision,recall,f1 score,roc auc score
from sklearn.metrics import roc_auc_score
yprob_a=knn.predict_proba(x_test)[:,1]
roc_auc=roc_auc_score(y_test,yprob_a)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
# given as list as works better
metric_df=pd.DataFrame({
    'Model':['KNN'],
    'Accuracy':[accuracy_score(y_test,y_pred)],
    'Precision':[precision_score(y_test,y_pred)],
    'Recall':[recall_score(y_test,y_pred)],
    'F1 score':[f1_score(y_test,y_pred)],
    'ROC AUC':[roc_auc]
})

metric_df
```

```python
metric_df.round(4)
```

# Hyperparameter Tuning

## What are Hyperparameters?

- Model parameters?
  - Learned from data during training
  - Change automatically when you call 'fit()'
  - Eg:
    - Logistic regression: coefficients w1,w1,intercept b
    - Neural networks: weights for all layers

- Hyper Parameters
  - **set by you,before training**
  - they control how the algorithm learns or behaves,ground rules for model to learn
  - they are not learned from the data,the algorithm treats them as fixed.
  - Example:
    - KNN: no.of neighbours was fixed(5) the algorithm shd only look for closest neighbours and not anybody else, and distance by defalut we said eucledian is a hyperparameter for euclidean there are other manhattan

## why do we need hyperparameters?

- different hyperparameters values can give different performance.(values)
- wrong hyperparameters---> underfitting or overfitting
  - Underfitting: Model generalises info from neighbors and does not calculate

## what are the methods of hyperparameter tuning(in ML terms)

- Underfitting:
  - model is too simple
  - Ignores local patterns---> leads to high bias and low variance
- Overfitting:
  - model is too complex and learns the training data's noise and performs well on the training data but poorly on test data.
  - model memorises the data and just gives it as it is ---> very low bias,high varaince

When is a model considered good?
- when the training data patterns are correctly understood by the model
- when it generalizes well to new,unseen data
- not memorize the training data(overfitting) or fail to capture relevant patterns(underfitting)

## How to implement it?

will search and tell if model good considering hyperparameters
1. Manual Search- not suggested as not systematic- cant do trail and error method manually
2. Grid Search(really exhaustive)- `GridSearchCV`

- values for each hyperparameter it will check and convert this to a grid
- the algorithm will test every single combination using cross- validation.

3. Random Search- `RandomizedSearchCV`

- you can define a range for each hyperparameter
- randomly sample combinations and evaluate them
- more effective

more like bayesian optimization

doing for knn the above dataset

```python
# and splitting dataset:into x and y,we split using the train-test split
# x--> scaled(important if in KNN)

# x_train,x_test,y_train and y_test
```

```python
# GridSearchCV

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

#define the parameter grid using weights and defining
param_grid={
    # 'hyperparameter': ['value1','value2','value3']
    'n_neighbors': [3,5,7,9,11], # we are giving specific values not range of values # a range of values in Randomized search
    'weights': ['uniform','distance'],
    'metric':['minkowski'], # error calculation
    'p':[1,2] # p=1--> manhattan,p=2---> eucledian
}
# leaf size
```

```python
grid_search=GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    scoring='accuracy', # how well optimized
    cv=5 # 5 fold cross validation
)
```

```python
# run the grid
grid_search.fit(x_train,y_train)

# best parameters
grid_search.best_params_
```

```python
# cv accuracy for this combination
grid_search.best_score_
```

```python
from sklearn.metrics import accuracy_score,classification_report

best_knn=grid_search.best_estimator_ # knn changed to grid search's best_estimator_ in that only training done so now only prediction with new knn

y_pred_best=best_knn.predict(x_test)

accuracy_best=accuracy_score(y_test,y_pred_best)
accuracy_best
```

```python
metric_df.loc[1]=[
    'Grid search',
    accuracy_score(y_test,y_pred_best),
    precision_score(y_test,y_pred_best),
    recall_score(y_test,y_pred_best),
    f1_score(y_test,y_pred_best),
    roc_auc_score(y_test,y_pred_best)]


metric_df

# increased
# this is called as hyperparameter tuning by setting
```
